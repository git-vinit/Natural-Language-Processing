```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')
```

```python
#DL
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Embedding
```

```python
df=pd.read_csv('twitter.csv')
```

```python
df
```

```python
stlist=stopwords.words("english")
lm=WordNetLemmatizer()
```

```python
def preprocess(text):
  text=text.lower()
  tokens=word_tokenize(text)
  clean_tokens=[t for t in tokens if t.isalpha()]
  clean_tokens_1=[t for t in clean_tokens if t not in stlist]
  lem_tokens=[lm.lemmatize(t) for t in clean_tokens_1]
  return ' '.join(lem_tokens)                
```

```python
df['clean_tweets']=df['tweet'].apply(preprocess)
```

```python
df
```

```python
# feature and target
x=df['clean_tweets']
y=df['label']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#keras tokenizer
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
```

```python
tok=Tokenizer()
tok.fit_on_texts(xtrain)
```

```python
voc=tok.index_word
vocab_len=len(voc)
vocab_len
```

```python
# sequence
train_seq=tok.texts_to_sequences(xtrain)
```

```python
print(train_seq)
```

```python
doc_len=[]
for i in train_seq:
  doc_len.append(len(i))
print(doc_len)
```

```python
max(doc_len)
```

```python
import numpy as np
np.quantile(doc_len,0.9)
```

```python
maxlenght=51
# padding
train_matrix=sequence.pad_sequences(train_seq,maxlen=maxlenght)
train_matrix
```

```python
#Testing data
test_seq=tok.texts_to_sequences(xtest)
test_matrix=sequence.pad_sequences(test_seq,maxlen=maxlenght)
test_matrix
```

```python
# model
model=Sequential()
model.add(Embedding(input_dim=vocab_len+1,
                    input_length=maxlenght,
                    output_dim=100,
                    mask_zero=True))
model.add(Flatten())
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
model.fit(train_matrix,ytrain,epochs=20,batch_size=32)
```

```python
#prediction
from sklearn.metrics import classification_report
ypred=model.predict(test_matrix)
ypred=np.where(ypred>=0.5,1,0)
print(classification_report(ytest,ypred))
```

```python
sms1="user father dysfunctional selfish drag kid dys"
sms2="yes it when you call michelleobama a gorilla because racists have long thought of black people as no betâ"
```

```python
def predict_sms(sms):
  dataseq=tok.texts_to_sequences([sms])
  datamatrix=sequence.pad_sequences(dataseq,maxlen=maxlenght)
  y_pred=model.predict(datamatrix)
  y_pred=np.where(y_pred>=0.5,1,0)
  output=le.inverse_transform(y_pred[0])[0]
  return output
```
